# Data Understanding and Cleaning


schedule.csv =
```
id	: ID Stasiun-Kereta, contoh: AC-2400 sampai YK-7102, terdapat 18065 kereta
station_id : ID Stasiun, contoh: AC, terdapat 100 id stasiun
train_id : ID Kereta, contoh: 2400
line : Daerah Commuter Line/ Wilayah Operasi, terdapat 10. contoh: Commuter Line TANJUNGPRIUK, CIKARANG, YK, BOGOR, TANGERANG, BST, RANGKASBITUNG, PRAMBANAN EKSPRES (PRAMEKS), PRAMBANAN EKSPRES, MERAK.
route	: Rute kereta, terdapat 69 rute berbeda. contoh: JAKARTAKOTA-TANJUNGPRIUK
color	: Warna kereta, terdapat 6 warna berbeda
destination	:  tujuan, terdapat 25 daerah tujuan berbeda.
time_estimated :  Departure- waktu brangkat
destination_time :	Arrival- waktu datng
updated_at : Last time update
```

---

station.csv :

```
id : ID Stasiun, contoh: AC, terdapat 100 id stasiun
name : Nama Stasiun, terdapat 100 nama stasiun
daop : Daerah OperasionalKAI, terdapat 3 jenis yaitu daop 1, 2, dan 6. 1 = Jakarta, 2= Bandung, 3 = cirebon, 4= semarang, 5= purwokerto, 6= jogja.
fg_enable : hanya bernilai 1
have_schedule	: Is the Station already make a schedule? TRUE or FALSE
updated_at : Last time update
**bold text**
```

## Trip Duration| Waktu Perjalanan

````python
import pandas as pd

# Assuming your DataFrame is named 'df'

# Convert time_estimated and destination_time to datetime format
df['time_estimated'] = pd.to_datetime(df['time_estimated'])
df['destination_time'] = pd.to_datetime(df['destination_time'])

# Menghitung durasi perjalanan dari stasiun awal ke stasiun tujuan dalam menit
df['trip_duration_minutes'] = (df['destination_time'] - df['time_estimated']).dt.total_seconds() / 60
df['trip_duration_minutes'] = df['trip_duration_minutes'].astype(int)
df.head()
````
